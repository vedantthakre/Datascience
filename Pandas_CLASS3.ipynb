```python
import pandas as pd

df4 = pd.DataFrame({"a": [3,4,5,2,1,3,4,5,6]})
print(df4)
```

```python
"""
What is Windowing Function?

👉 It means:
Apply calculation on a group (window) of nearby values

🔹 Simple Idea 🧠

👉 Instead of using whole column
👉 Work on small part (window)

1️⃣ Rolling Window
df['a'].rolling(3).mean()

👉 Window size = 3
👉 Take 3 values at a time

(3,4,5) → mean = 4
(4,5,2) → mean = 3.66
(5,2,1) → mean = 2.66
...
"""
```

```python
df4['a'].rolling(3).mean()
```

```python
df4['a'].rolling(3).sum()
```
