```python
data1 = pd.DataFrame({
    'Key1' : [3,4,5,6,7],
    'Key2' : [6,6,5,6,7],
    'Key3' : [8,49,5,6,7],
})

data1
```

```python
data2 = pd.DataFrame({
    'Key1' : [3,54,5,76,87],
    'Key4' : [6,63,5,56,7],
    'Key5' : [58,449,5,556,57],
})

data2
```

```python
pd.merge(data1,data2) #An inner merge (also called inner join) means:👉 Keep only the common (matching) rows from both datasets

```

```python
pd.merge(data1,data2,how = 'left')
```

```python
pd.merge(data1,data2,how = 'right')
```

```python
pd.merge(data1,data2,how = 'cross')
```

```python
pd.merge(data1,data2,how = 'outer' ,on = 'Key1')
```

```python
import pandas as pd  # 👉 .join() works on index, not on columns like merge()

data1 = {'Name': ['A', 'B', 'C']}
data2 = {'Marks': [80, 90, 70]}

df1 = pd.DataFrame(data1, index=['a', 'b', 'c'])
df2 = pd.DataFrame(data2, index=['b', 'c', 'd'])

df1.join(df2, how='right')
```

```python
df1.join(df2, how='inner')
```

```python
def fare_category(x):
    if x < 10:
        return 'cheap'
    elif 10 <= x <= 20:
        return 'mid'
    else:
        return 'high'

df['category'] = df['Fare'].apply(fare_category)
```
